*DataMol is a powerful tool that allows us to perform molecular descriptor calculations for large sets of compounds. These descriptors are numerical values that summarize important properties of a molecule, such as its size, shape, and chemical reactivity. They can be used for various purposes, such as virtual screening, drug design, and quantitative structure-activity relationship (QSAR) modeling.*

*Molecular descriptors are essential in drug discovery and design. By using them, we can analyze and compare the properties of large sets of compounds, identify promising drug candidates, and optimize their properties. This can save a lot of time and resources compared to traditional trial-and-error methods. Moreover, molecular descriptors can also help us understand the structure-activity relationships of compounds, which is crucial for designing effective drugs.requested_image.png*

```python
#@title ðŸ”¢ðŸš€ **let's start the calculation!**
import pandas as pd
from rdkit import Chem
from mordred import Calculator, descriptors

# Cargar los SMILES del archivo naturar.smi
with open('naturar.smi', 'r') as f:
    smiles = f.readlines()
    smiles = [smi.strip() for smi in smiles]

# Calcular todos los descriptores para todas las molÃ©culas al mismo tiempo
calc = Calculator(descriptors)
mols = [Chem.MolFromSmiles(smi) for smi in smiles]
desc_matrix = calc.pandas(mols)

# Agregar los SMILES como una columna del dataframe
desc_matrix['SMILES'] = smiles

# Visualizar los resultados
print(desc_matrix.head())

# Exportar los resultados a un archivo CSV
desc_matrix.to_csv('resultados.csv', index=False)

# Imprimir una confirmaciÃ³n de que se ha guardado el archivo
print('Los resultados se han exportado a un archivo CSV llamado "resultados.csv".')
```
